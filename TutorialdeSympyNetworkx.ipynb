$$\Large \color{#C5A3A3}{\text{Matrices con Sympy}}$$

Realizaremos un tutorial para aprender a representar matrices, realizar operaciones algebraicas (como multiplicación y potencias) y obtener sus valores y vectores propios de forma exacta.

Para empezar, importamos las librerías necesarias.


```python
from sympy import Matrix, symbols, eye, diag, init_printing

# Mejora la visualización de las matrices en el Notebook
init_printing(use_unicode=True)
```

$$\color{#C5A3A3}{\text{Representar matrices y vectores}}$$

Para representar una matriz en SymPy, utilizamos el objeto Matrix.
Para la matriz tenemos que escribir sus filas como listas, y el vector debe escribirse como lista, de la siguiente manera:


```python
# Matriz de 3x2
A = Matrix([[1, -1], [3, 4], [0, 2]])

# Vector columna de 3 elementos
v = Matrix([1, 2, 3])

print("Matriz A:")
display(A)
print("Vector v:")
display(v)
```

$$\color{#C5A3A3}{\text{Obtener entradas, renglones y columnas}}$$



Para obtener una $\color{#C5A3A3}{\text{entrada}}$ particular, se utiliza M[i, j], donde i es la fila y j la columna, es necesario recordar que en Python se inicia en 0.

```python
# Entrada específica: A[fila, columna]
A[0, 1]
```

Para obtener una $\color{#C5A3A3}{\text{fila/columna}}$ completa, se utilizan los métodos .row(i) y .col(j)

```python
# Fila completa
fila = A.row(0)
fila
```

```python
# Columna completa
columna = A.col(1)
columna
```

$$\color{#C5A3A3}{\text{Multiplicación de matrices}}$$

Creamos matrices con dimensiones compatibles

```python
A = Matrix([[1, 2], [3, 4]])
B = Matrix([[0, 1], [1, 0]])

# Multiplicación de matrices
C = A * B

C
```

$$\color{#C5A3A3}{\text{Multiplicación de matriz por un vector}}$$


```python
# Creamos un vector columna
vector = Matrix([1, 2])

# Multiplicar un vector por una matriz
prod_vector = A * vector
prod_vector
```

$$\color{#C5A3A3}{\text{Elevar una matriz a una potencia}}$$

Para elevar una matriz a una potencia, se utiliza el operador de doble **, esto es equivalente a hacer A x A

```python
# Elevar a una potencia
potencia = A**2

potencia
```

$$\color{#C5A3A3}{\text{Valores y Vectores Propios}}$$

El método .eigenvals() devuelve los valores propios y .eigenvects() nos da una lista que incluye los valores propios, sus multiplicidades y los vectores propios asociados.

```python
# Valor propio
valores_p = A.eigenvals()
valores_p
```

```python
# Vectores propios
vectores_p = A.eigenvects()
vectores_p
```

$$\Large \color{#C5A3A3}{\text{Diagramas de transición con NetworkX}}$$

Un diagrama de transición se define mediante nodos, que representan los estados, y aristas dirigidas, que representan las probabilidades de cambio entre ellos.

Primero importamos las librerías necesarias.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

$$\color{#C5A3A3}{\text{Creación del Grafo Dirigido}}$$

Para representar procesos donde el sentido de la transición importa (como de un estado A a un estado B), utilizamos la clase DiGraph

```python
# Crear un objeto de grafo dirigido
G = nx.DiGraph()
```

$$\color{#C5A3A3}{\text{Nodos}}$$

Como ya se menciono los nodos representan los estados, los definimos de la siguiente manera

```python
# Definir los estados (nodos)
G.add_nodes_from(["1", "2", "3"])
```

$$\color{#C5A3A3}{\text{Transiciones (Aristas)}}$$

Las aristas nos indican hacia donde se dirige, weight nos indica la probabilidad

```python
# Añadir aristas con sus respectivas probabilidades
# Estructura: (Origen, Destino, Probabilidad)
G.add_edge("1", "2", weight=0.7)
G.add_edge("2", "3", weight=0.5)
G.add_edge("3", "1", weight=1.0)
G.add_edge("2", "2", weight=0.5)
```

$$\color{#C5A3A3}{\text{Diagrama}}$$

Una vez que tenemos el grafo, los nodos y las aristas, dibujamos el diagrama

```python
# Definir la disposición de los nodos
pos = nx.spring_layout(G)

# Dibujar nodos y aristas
nx.draw(G, pos, with_labels=True, node_color='#C5A3A3', node_size=1500)

# Probabilidades
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

$$\Large \color{#C5A3A3}{\text{Ejemplo}}$$



$$\color{#C5A3A3}{\text{Definimos la Matriz de transición}}$$

Primero, definimos la Matriz de Transición, la suma de cada renglón debe ser 1 (100%).

```python
# 1. Definimos la matriz de transición P (Cada renglón suma 1)
P = Matrix([[0.8, 0.2],
            [0.4, 0.6]])
```

$$\color{#C5A3A3}{\text{Creamos el grafo}}$$


```python
G = nx.DiGraph()
```

$$\color{#C5A3A3}{\text{Agregamos los estados}}$$

```python
G.add_nodes_from(["E1", "E2"])
```

$$\color{#C5A3A3}{\text{Agregamos las transiciones}}$$

```python
G.add_edge("E1", "E1", weight=0.8)
G.add_edge("E1", "E2", weight=0.2)
G.add_edge("E2", "E1", weight=0.4)
G.add_edge("E2", "E2", weight=0.6)
```

$$\color{#C5A3A3}{\text{Dibujamos el diagrama}}$$

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='#C5A3A3', node_size=2000, arrowsize=20)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.title("Diagrama de Transición")
plt.show()
```

$$\color{#C5A3A3}{\text{Explicación}}$$

Descripción del modelo:

* Si estamos en $E_1$, hay un 80% de probabilidad de quedarse ahí y 20% de pasar a $E_2$.

* Si estamos en $E_2$, hay un 40% de probabilidad de pasar a $E_1$ y 60% de quedarse ahí.
