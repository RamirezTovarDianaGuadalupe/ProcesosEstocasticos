$\Large \color{#C5A3A3}{\text{1. Distribuciones Discretas}}$

---

Estas distribuciones describen variables aleatorias que toman valores en conjuntos finitos o infinitos numerables.
##

$\color{#C5A3A3}{\text{1.1. Distribución Bernoulli}}$

Se utiliza para un solo experimento con dos resultados posibles: éxito ($p$) o fracaso ($1-p$).
* **Notación:** $X \sim \text{Bernoulli}(p)$
* **Función de probabilidad:** $P(X=x)=p^{x}(1-p)^{1-x}, \quad x\in\{0,1\}$
* **Función acumulada:** $$F(x)=\begin{cases}0 & x<0 \\ 1-p & 0\le x<1 \\ 1 & x\ge1 \end{cases}$$
* **Momentos:** $E[X]=p, \quad Var(X)=p(1-p)$
* **FGM:** $M_{X}(t)=(1-p)+pe^{t}$


$\color{#C5A3A3}{\text{1.2. Distribución Binomial}}$
##

Representa el número de éxitos en $n$ ensayos independientes de Bernoulli.
* **Notación:** $X \sim \text{Bin}(n,p)$
* **Función de probabilidad:** $P(X=x)=\binom{n}{x}p^{x}(1-p)^{n-x}, \quad x=0,1,...,n$
* **Momentos:** $E[X]=np, \quad Var(X)=np(1-p)$
* **FGM:** $M_{X}(t)=(1-p+pe^{t})^{n}$

$\color{#C5A3A3}{\text{1.3. Distribución Poisson}}$
##

Modela el número de eventos en un intervalo continuo de tiempo o espacio.
* **Notación:** $X \sim \text{Poisson}(\lambda)$
* **Función de probabilidad:** $P(X=x)=\frac{e^{-\lambda}\lambda^{x}}{x!}, \quad x=0,1,2,...$
* **Momentos:** $E[X]=\lambda, \quad Var(X)=\lambda$
* **FGM:** $M_{X}(t)=\exp(\lambda(e^{t}-1))$

$\Large \color{#C5A3A3}{\text{2. Distribuciones Continuas}}$

---

Se usan para variables que pueden tomar cualquier valor en un intervalo real.

$\color{#C5A3A3}{\text{2.1. Distribución Uniforme}}$
Indica que todos los intervalos de igual longitud en el soporte tienen la misma probabilidad.
* **Notación:** $X \sim U(a,b)$
* **Densidad:** $$f(x)=\begin{cases}\frac{1}{b-a} & a\le x\le b \\ 0 & \text{otro caso}\end{cases}$$
* **Acumulada:** $$F(x)=\begin{cases}0 & x<a \\ \frac{x-a}{b-a} & a\le x\le b \\ 1 & x>b\end{cases}$$
* **Momentos:** $E[X]=\frac{a+b}{2}, \quad Var(X)=\frac{(b-a)^{2}}{12}$
* **FGM:** $M_{X}(t)=\frac{e^{tb}-e^{ta}}{t(b-a)}$



$\color{#C5A3A3}{\text{Distribución Exponencial}}$

Común para modelar tiempos de espera entre eventos de un proceso de Poisson.
* **Notación:** $X \sim \text{Exp}(\lambda)$
* **Densidad:** $f(x)=\lambda e^{-\lambda x}, \quad x\ge0$
* **Acumulada:** $F(x)=1-e^{-\lambda x}$
* **Momentos:** $E[X]=\frac{1}{\lambda}, \quad Var(X)=\frac{1}{\lambda^{2}}$
* **FGM:** $M_{X}(t)=\frac{\lambda}{\lambda-t}, \quad t<\lambda$

$\color{#C5A3A3}{\text{2.3. Distribución Normal}}$

Distribución simétrica fundamental por el Teorema del Límite Central.
* **Notación:** $X \sim N(\mu, \sigma^{2})$
* **Densidad:** $f(x)=\frac{1}{\sqrt{2\pi\sigma^{2}}}\exp(-\frac{(x-\mu)^{2}}{2\sigma^{2}})$
* **Momentos:** $E[X]=\mu, \quad Var(X)=\sigma^{2}$
* **FGM:** $M_{X}(t)=\exp(\mu t+\frac{1}{2}\sigma^{2}t^{2})$


$\color{#C5A3A3}{\text{2.4. Distribución Gamma}}$

Generalización de la exponencial para modelar la suma de tiempos de espera.
* **Notación:** $X \sim \text{Gamma}(\alpha, \beta)$
* **Densidad:** $f(x)=\frac{\beta^{\alpha}x^{\alpha-1}e^{-\beta x}}{\Gamma(\alpha)}, \quad x>0$
* **Momentos:** $E[X]=\frac{\alpha}{\beta}, \quad Var(X)=\frac{\alpha}{\beta^{2}}$
* **FGM:** $M_{X}(t)=(\frac{\beta}{\beta-t})^{\alpha}, \quad t<\beta$
